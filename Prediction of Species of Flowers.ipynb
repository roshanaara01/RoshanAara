**DATA PROCESSING**

```python
data = data.sample(frac=1)
```

```python
new_data = data.set_index('Id')
```

```python
from sklearn import preprocessing
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data['Encoded_species'] = LabelEncoder().fit_transform(data['Species'])
```

```python
new_data.head(15)
```

```python
encoded_array = LabelEncoder().fit_transform(new_data['Species'])
```

```python
encoded_array
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = new_data.iloc[:,:-2].values
```

```python
x
```

```python
x.shape
```

```python
y = encoded_array
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

**Building the Decision tree model, training and predictions**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(x_train, y_train)
```

```python
predictions = dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

**Testing the accuracy of the data**

```python
print(classification_report(y_test, predictions))
```

**WE GOT 96% ACCURACY FOR THE DATASET**

```python
print(confusion_matrix(y_test, predictions))
```

**Visualizing the decision tree using Graphviz**

```python
from IPython.display import Image
from six import StringIO
from sklearn.tree import export_graphviz
```

```python
import pydotplus
```

```python
features = list(new_data.columns[:-1])
features
```

```python
target = list(new_data.Species.unique())
target
```

```python
dot_data = StringIO()
export_graphviz(dtree, out_file= dot_data, feature_names=features, filled= True, rounded= True, special_characters= True, class_names=target)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

**Decision tree using sklearn**
